```python
import pandas as pd

with open('yelp_academic_dataset_review.json', 'rb') as f:
    data = f.readlines()[:100]
    data = map(lambda x: x.rstrip(), data)
    data_json_str = "[" + ','.join(data) + "]"
    data_df = pd.read_json(data_json_str)
```

```python
data_df.head(10)
```
