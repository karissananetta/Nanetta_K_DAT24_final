# In order to improve the trendsetter prediction, some feature engineering is required. The features aim to answer the following questions:

## SPEED: How early does a user leave reviews for businesses, on average?
* Feature engineering 1: Find the average age of an establishment when it is reviewed by a particular user (review date - first/min review date)
* Feaure engineering 2: Find how many reviews have been left for each of the businesses before this user leaves a review

## QUALITY: How well are the businesses (that receive reviews from the users) doing?
* Feature engineering 3: What is the average star ratings of the businesses that a particular user reviews?
* Feature engineering 4: How many reviews do the businesses have to date, on average?

```python
full = data.join(first_rating_date, on='business_id', how='left', lsuffix='_original', rsuffix='_min')
```

```python
full.head()
```

```python
full = full.drop(full.columns[[0]], axis=1)
```

```python
full.head()
```

```python
full['date_original'] = pd.to_datetime(full['date_original'])
```

```python
full['date_min'] = pd.to_datetime(full['date_min'])
```

```python
full['review_age'] = full['date_original'] - full['date_min']
```

```python
# x.item gives the number of nanoseconds from np.timedelta64
# then convert the nanoseconds to days... -_-

full['review_age_days'] = full['review_age'].apply(lambda x: x.item() / (1000000000 * 60 * 60 * 24))
```

```python
group_by_user = full.groupby('user_id')
mean_review_age_days_by_user = group_by_user['review_age_days'].mean()
mean_review_age_days_by_user.head()
```

```python
full2 = full.join(mean_review_age_days_by_user, on='user_id', how='left', lsuffix='_original', rsuffix='_mean')
```

```python
# for each row in full2, get the number of reviews that have been left before a particular user leaves a review
# for each row, count reviews < date_original

business_id_to_review_dates = {}
for _, row in full2.iterrows():
    if row['business_id'] in business_id_to_review_dates:
        business_id_to_review_dates[row['business_id']].append(row['date_original'].value / 10**9)
    else:
        business_id_to_review_dates[row['business_id']] = [row['date_original'].value / 10**9]

```

```python
# Sort each list
for b_id in business_id_to_review_dates:
    business_id_to_review_dates[b_id].sort()
```

# Feature engineering 4: How many reviews do the businesses have on average for each user? (to determine popularity of businesses)

```python
# Step 1: use the business dataset to get the review count for each business
# Step 2: use the review dataset to append the review count based on the business_id
# Step 3: groupby based on user_id and get the average review count for each user
# Step 4: append the average review count to the full3b dataset, forming full4 dataframe
```

```python
# Step 1: use the business dataset to get the review count for each business

business_review_count = pd.read_csv('yelp_academic_dataset_business.csv')
business_review_count = business_review_count[['business_id', 'review_count']]
business_review_count.head()
```

```python
# Step 2: use the review dataset to append the review count based on the business_id

business_and_review_count_data = pd.merge(full3b, business_review_count, how='left', on='business_id')
business_and_review_count_data.head()
```

```python
# Step 3: groupby based on user_id and get the average review count for each user

grouped_fe4 = business_and_review_count_data.groupby('user_id')
fe4 = grouped_fe4['review_count'].mean()
fe4.head()
```
